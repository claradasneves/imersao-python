```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplfinance as mpf
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# O yf é uma biblioteca que conecta dados de ações de empresas com o código
# Coloca o nome da ação, .SA que signifca que é br e a data de começo e fim

dados = yf.download('PETR4.SA', start='2023-01-01', end='2023-12-31')
dados
```

```python
dados.columns = ['Abertura', 'Maximo', 'Minimo', 'Fechamento', 'Fech_ajust', 'Volume']
dados =  dados.rename_axis('Data')
dados
```

```python
dados['Fechamento'].plot(figsize=(10,6))
plt.title('Varição do preço por data', fontsize=16)
plt.legend('Fechamento')
```

```python
 # Só as 60 primeiras linhas
df = dados.head(60).copy()

# Nova coluna que vai pegar o índice
df['Data'] = df.index

# Converte datas pro formato numérico do matplotlib
# Assim, ele consegue printar as datas certinhos no gráfico
df['Data'] = df['Data'].apply(mdates.date2num)
df
```

```python
fig, ax = plt.subplots(figsize=(15,8))

# Largura dos candle
width =  0.7

for i in range(len(df)):
  # Cor do cande sendo definida
  # se Fechamento >  Abertura, cande é verde
  # se Fechamento <  Abertura, cande é vermelho
  if df['Fechamento'].iloc[i] > df['Abertura'].iloc[i]:
    color = 'green'
  else:
    color = 'red'

  # Desenhando alinha vertical do candle (mecha)
  # Mostra ps preços máximo (topo da linha) e mínimo (base da linha) do dia
  # Usamos 'ax.plot' para desenhar uma linha vertical

  ax.plot([df['Data'].iloc[i], df['Data'].iloc[i]],
          [df['Minimo'].iloc[i], df['Maximo'].iloc[i]],
          color = color,
          linewidth=1)

  ax.add_patch(plt.Rectangle((df['Data'].iloc[i] - width/2, min(df['Abertura'].iloc[i], df['Fechamento'].iloc[i])),
                             width,
                             abs(df['Fechamento'].iloc[i] - df['Abertura'].iloc[i]),
                             facecolor=color))

df['MA7'] = df['Fechamento'].rolling(window=7).mean()
df['MA14'] =  df['Fechamento'].rolling(window=14).mean()

# Plotando as médias móveis
ax.plot(df['Data'], df['MA7'], color='orange', label='Média Móvel 7 Dias')
ax.plot(df['Data'], df['MA14'], color='yellow', label='Média Móvel 14 Dias')

# Adicionando legendas para as médias móveis
ax.legend()

# Formatar os eixos
```
