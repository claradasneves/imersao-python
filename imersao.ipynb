Aula 03 -  Manipulação de Dados e Criação de Gráficos


```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
# modificando os nomes das colunas
df_principal = df_principal.rename(columns={'Ativo':'ativo', 'Data': 'data', 'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal

# a função .copy() apaga o original e cria um novo, sem ter relações
```

```python
# criando novas coluna
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['var_pct'] + 1)
df_principal
```

```python
# fazendo o merge, dá pra fazer só com o left, mas ele assume que os nomes são iguais
# na última parte, o "how" serve para ver qual mantém no principal
df_principal = df_principal.merge(df_total_acoes, left_on='ativo', right_on='Código', how='left')

# curiosidade: se eu ficar rodando isso ele vai criando colunas de "Código_x", "Código_y" etc

# drop col. Código
# DÚVIDA: o .copy() não é necessário agora?
df_principal = df_principal.drop(columns=['Código'])

# visualização
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica': 'qtd_acoes'}).copy()
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['qtd_acoes']

# visualizacao
df_principal
```

```python
# formatar todos os dados do para tipo float
pd.options.display.float_format = '{:.2f}'.format
```

```python
# convertendo especificamente a parte da qtd_teorica para int usando .astype()
df_principal['qtd_acoes'] = df_principal['qtd_acoes'].astype(int)

df_principal
```

```python
# if df_principal['varicao_rs'] > 0:
#   df_principal['resultado'] = 'Subiu'
# elif df_principal['varicao_rs'] < 0:
#   df_principal['resultado'] = 'Desceu'
# else:
#   df_principal['resultado'] = 'Estável'

# não pode fazer isso que eu fiz lá em cima........


# funcao apply: é chamada para aplicar outra função
# lambda: faz operações de linha a linha
df_principal['resultado'] = df_principal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else('Desceu' if x < 0 else 'Estável'))

df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])

df_principal
```

```python
df_chatgpt = pd.read_excel("/content/acoes_tabela.xlsx", sheet_name="ChatGPT")
df_chatgpt
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Empresa', how='left')
df_principal = df_principal.drop(columns=['Empresa'])

df_principal = df_principal.rename(columns={'Ano de Fundação':'ano_fundacao'}).copy()
```

```python
maior = df_principal['variacao_rs'].max()

menor = df_principal['variacao_rs'].min()

media = df_principal['variacao_rs'].mean()

media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['variacao_rs'].mean()

media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['variacao_rs'].mean()

print(f"Maior\tR$ {maior: ,.2f}")
print(f"Menor\tR$ {menor: ,.2f}")
print(f"Média\tR$ {media: ,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu: ,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu: ,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']

df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('resultado')['variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
# Função que transforma o valor para o formato de reais

def formatar_moeda(valor):
  return 'R$ {: .2f}'.format(valor)
```

```python
df_analise_saldo['variacao_rs_formatada'] =  df_analise_saldo['variacao_rs'].map(formatar_moeda)

df_analise_saldo
```

```python
# Antes estava dando errado porque ele considerava o variacao_rs_formatada no y. Mude para o variacao_rs

fig_saldo_resultado = px.bar(df_analise_saldo, x = 'resultado', y = 'variacao_rs', text = 'variacao_rs_formatada', title = 'Variação Reais por Resultado', labels={'Resultado': 'Resultado', 'Variacao_Real': 'Variação Real'})
fig_saldo_resultado.show()
```

```python
fig_analise_segmento = px.pie(df_analise_segmento, values='variacao_rs', names='Segmento', title='Variação em R$ por segmento')
fig_analise_segmento.show()

# Problema: outliers
```
